```python
# filling missing vlaue

df.fillna(0)
```

```python
# filling missng value with menan of coulumn
df['Salry-fillna']=df['Salary'].fillna(df['Salary'].mean())
```

```python
df
```

```python
# renaming columns\\
df=df.rename(columns={'Joining_Date':'Date_of_joining'})
df.head()
```

```python
# chage data type
df['Salry-fillna_float']=df['Salry-fillna'].astype(float)
df.head()
```

```python
df.dtypes
```

```python
# increase salary by 10%
df['salary_incre_10%']=df['Salary'].apply(lambda x:x+x*0.1)
df
```

```python
# data aggregating and grouping

df.head()
```

```python
grouped_mean=df.groupby('Department')['ID'].count()
grouped_mean
```

```python

grouped_mean_name=df.groupby('Name')['Salary'].mean()
grouped_mean_name
```

```python
df['Name'].unique()
```

```python
## aggregae multiple functions

grouped_aggregration=df.groupby('Department')['Salary'].agg(['mean','sum','count'])
grouped_aggregration
```

```python
### merging and joining dataframe

df1 = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "Finance", "IT", "Sales", "Marketing"]
})

# Second DataFrame: Employee salary details
df2 = pd.DataFrame({
    "Emp_ID": [101, 102, 104, 106],
    "Salary": [50000, 60000, 75000, 80000],
    "Bonus": [5000, 7000, 8000, 9000]
})

merged_df = pd.merge(df1, df2, on="Emp_ID", how="inner")
print(merged_df)
```

```python
merged_df = pd.merge(df1, df2, on="Emp_ID", how="outer")
print(merged_df)
```

```python
merged_df = pd.merge(df1, df2, on="Emp_ID", how="left")
print(merged_df)
```

```python
merged_df = pd.merge(df1, df2, on="Emp_ID", how="right")
print(merged_df)
```
