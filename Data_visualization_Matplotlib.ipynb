# MatPlot Lib

#### histograms are used to represent the distribution of a datasets. They divide the data into bins and count the number of data point in each bin

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
# create a line plot

plt.figure()
plt.title("number and square")
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# create customize line plot
plt.plot(x,y,color="r",linestyle='--',marker='*',markersize=50)

plt.grid()
```

```python
# create multiple plot

# sample data
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x,y)


plt.subplot(2,2,2)
plt.plot(x,z,color="r")

plt.subplot(2,2,3)
plt.plot(z,x,color='b')

plt.subplot(2,2,4)
plt.plot(y,z,color='#00FF00')
```

```python
# Bar plot
category=['a','b','c','d','e']
values=[12,365,26,85,59]

plt.bar(category,values, color="#00FF00")
```

```python
# hisotgram

data=[1,2,3,2,3,2,1,4,5,5,4,1,2,5,4,1,2,3,5,4]
plt.hist(data,bins=5,color="orange",edgecolor="b")
```

```python
# scatter plot

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.scatter(x,y,marker="*")
```

```python
# pie chart
labels=['A','B','C','D','E']
size=[30,20,40,10,25]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,1,0)  # to move out the slice

# create a pie chart

plt.pie(size,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```
