# Pandas Dataframe and Series

### DataFrame is two dimentional ,size mutable and potentially heterogeneous tabulat data structure with labelled axes

### Series  is a one dimentional array like object


```python
import pandas as pd
```

```python
## Series

data=[1,2,3,4,5]
series=pd.Series(data)
print("series is : \n",series)
```

```python
# create series from dectionary 
data1={'a':1,'b':2,'c':6}
series1= pd.Series(data)
print(series1)
```

```python
# create our own indexs
data2=[10,20,30]
index1=['a','b','c']
series2=pd.Series(data2,index=index1)
print(series2)
```

```python
# DataFrame
# create datafrme from dictionary of list
data3={
    'name':['siddhesh','santosh','sagar'],
    'age':[22,55,33],
    'city':['chembur','mumbai','kupavade']
}
dataframe=pd.DataFrame(data3)
print(dataframe)
```

```python
# create datafrmae form list of dictionary
data4=[
    {'name':'siddhesh','age':22,'city':'mumbai'},
    {'name':'mahesh','age':12,'city':'mumbai'},
{'name':'rohan','age':2,'city':'mumbai'}
]

dataframe1=pd.DataFrame(data4)
print(dataframe1)
```
