# reading data from different data source


```python
# first convert json data into string using StringIO  and then read it using pandas

# to convert df into json data use [pd.to_json("file name")



import pandas as pd
from io import StringIO

# Correct JSON string
data = """
[
  {
    "Emp_ID": 101,
    "Name": "Alice",
    "Department": "HR",
    "Salary": 50000,
    "Bonus": 5000
  },
  {
    "Emp_ID": 102,
    "Name": "Bob",
    "Department": "Finance",
    "Salary": 60000,
    "Bonus": 7000
  },
  {
    "Emp_ID": 103,
    "Name": "Charlie",
    "Department": "IT",
    "Salary": 70000,
    "Bonus": 6000
  },
  {
    "Emp_ID": 104,
    "Name": "David",
    "Department": "Sales",
    "Salary": 75000,
    "Bonus": 8000
  },
  {
    "Emp_ID": 105,
    "Name": "Eva",
    "Department": "Marketing",
    "Salary": 65000,
    "Bonus": 5500
  }
]
"""

# Load JSON into pandas DataFrame
df = pd.read_json(StringIO(data))

print(df)
```
